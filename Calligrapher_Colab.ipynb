```
# 데모 실행
from demo import Hand

# 손글씨 생성
lines = [
    "Hello from Google Colab!",
    "This is handwriting synthesis",
]
biases = [0.75 for i in lines]
styles = [9 for i in lines]

hand = Hand()
hand.write(
    filename='output.svg',
    lines=lines,
    biases=biases,
    styles=styles
)

print("생성 완료! output.svg 파일을 확인하세요.")
```
